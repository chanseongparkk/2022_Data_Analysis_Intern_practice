### 1. 필요 라이브러리 import

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

### 2. 데이터셋 준비

```python
data = {'이름':["홍길동","성춘향","이몽룡","김철수","김영희","김만수","오지랖",
              "만순이","문정억","백만원","천만원","수십억","현재미"],
        "몸무게": [95, 85, 75, 70, 65, 55, 120, 100, 71, 65, 75, 100, 77],
        "키": [170, 170, 170, 170, 170, 175, 171, 165, 164, 177, 163, 192, 182],
        "거주지" : ["서울","대구","대전","서울","경북","서울","경북","서울",
                 "서울","전남","전북","경북","서울"],
        "주민번호": ["81XXXX-1XXXXXX", "81XXXX-1XXXXXX", "79XXXX-2XXXXXX", "71XXXX-1XXXXXXX",
                "65XXXX-2XXXXXXX", "81XXXX-1XXXXXX", "63XXXX-1XXXXXX", "81XXXX-3XXXXXX",
                "81XXXX-1234562", "55XXXX-1234562", "71XXXX-2233456", "92XXXX-1234562",
                "85XXXX-2234562"],
        "흡연여부": ["흡연", "비흡연", "비흡연", "흡연", "흡연", "비흡연", "흡연", "흡연", "흡연",
                 "비흡연","비흡연", "비흡연", "비흡연"]}

df = pd.DataFrame(data)
```

```python
# 데이터셋 확인
df.head()
```

```python
# 수정할 데이터프레임 복사

df_mod = df.copy()
```

### 3. 파생변수 생성

#### 3-1 이름, 흡연 여부(라벨인코딩 변수 생성)

```python
# 라벨 인코더 생성

encoding = LabelEncoder()

# 라벨인코딩
df_mod["이름_인코딩"] = encoding.fit_transform(df_mod["이름"])
```

```python
print(list(encoding.classes_)) #클래스 확인
print(list(encoding.inverse_transform([1,0]))) # 인코딩값으로 문자값 확인
```

```python
encoding2 = LabelEncoder()

# 라벨인코딩
df_mod["흡연여부_인코딩"] = encoding2.fit_transform(df_mod["흡연여부"])
```

```python
df_mod.head()
```

#### 3-2 나이 변수 생성

```python
# 주민번호로 나이 계산 함수 정의

year = 2022

def age(rnn):
  if int(rnn[:2]) < 21 and int(rnn[7]) in (3, 4) :
    birthyear = 2000 + int(rnn[:2])
  else:
    birthyear = 1900 + int(rnn[:2])
  return year - birthyear
```

```python
df_mod["나이"] = df_mod["주민번호"].apply(age)
```

```python
df_mod.head()
```

#### 3-3 성별 변수 생성

```python
# 성별 계산 함수 정의

def sex(rnn):
  if int(rnn[7]) == 1 or int(rnn[7]) == 3 :
    gen = "남"
  else:
    gen = "여"
  return gen
```

```python
df_mod["성별"] = df_mod["주민번호"].apply(sex)
```

```python
# 성별 onehotencoding

ohe = OneHotEncoder(sparse = False)

# 가변환된 값을 X_train_one으로 저장

df_mod_one = pd.DataFrame(ohe.fit_transform(df_mod[["성별"]]),
                          columns = ["성별_남", "성별_여"])
```

```python
df_mod = pd.concat([df_mod,df_mod_one],axis = 1)
```

```python
df_mod.head()
```

#### 3-4 나이 변수의 범주화 변수 생성

```python
df_mod["나이_범주"] = pd.cut(df_mod.나이, bins = [0,10,20,30,40,50,60,70],
                         labels = [0,10,20,30,40,50,60])
```

```python
df_mod.head()
```

#### 3-5 BMI 변수 생성

- BMI: 체질량지수, 인간의 비만도를 나타내는 지수로 체중과 키의 관계로 계산
- eval() 함수 이용: 문자열로 된 수식을 받아 그 결과를 return하는 함수

```python
# BMI 계산
df_mod.eval("BMI = 몸무게/ (키/100)**2", inplace = True)
```

```python
df_mod.head()
```

```python
# BMI 분류
df_mod["BMI분류"] = pd.cut(df_mod.BMI, bins = [0,18,23,25,30,90],
                          labels = ["저체중", "정상", "과체중", "비만", "고도비만"])

df_mod["BMI분류코드"] = df_mod["BMI분류"].map({"저체중":0, "정상":1, "과체중":2, "비만":3, "고도비만":4})
```
